```python
import plotly.graph_objects as go

i = 1
pc = samples[i]

def visualize_point_cloud(pc):
	fig = go.Figure(data=[go.Scatter3d(
	    x=pc[:, 0],
	    y=pc[:, 1],
	    z=pc[:, 2],
	    mode='markers',
	    marker=dict(
	        size=2,
	        #color=samples[i][:, 2],  # color points by z-axis value
	        #colorscale='Viridis',
	        opacity=0.8
	    )
	)])
	
	fig.update_layout(scene=dict(
	    xaxis_title='X',
	    yaxis_title='Y',
	    zaxis_title='Z'
	))
	
	fig.show()

visualize_point_cloud(pc)
```
